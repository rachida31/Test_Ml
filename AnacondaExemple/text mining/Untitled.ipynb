```python
from nltk.corpus import wordnet as wn#faire le sens des mots
```

```python
syn = wn.synsets('pain')
print("definition : ", syn[0].definition())#afficher la definition de mot pain
print("example : ", syn[0].examples())#afficher exemple avec le mot pain
```

```python
syn = wn.synsets('python')
print("definition : ", syn[0].definition())
```

```python
#faire la liste des synonymes
synonyms = []
for syn in wn.synsets ('Computer'):
    for lemma in syn.lemmas():
        synonyms.append(lemma.name())
        
print(synonyms)
        
```

```python
#faire la liste des antonymes
antonyms = []
for syn in wn.synsets ('small'):
    for lemma in syn.lemmas():
        if lemma.antonyms():
            antonyms.append(lemma.antonyms()[0].name())                
print(antonyms)
```

```python
# selection of stop words  (for this i use nltk.download('stopwords') to download a stop words)
from nltk.tokenize import sent_tokenize, word_tokenize # it's already imported
from nltk.corpus import stopwords # it's already imported
data =  "all persenne like me i like him."
stopWords = set(stopwords.words('english'))
words = word_tokenize(data)

ListStopWordSent = []
for w in words:
    if w   in stopWords: # if you went to select the important words, do you use if w not in stopWords.
        ListStopWordSent.append(w)
        
print(ListStopWordSent)


```

```python
from nltk.stem import WordNetLemmatizer # to display the verb or noun or adjective or adverd for a word 
lemmantizer = WordNetLemmatizer()
print(lemmantizer.lemmatize('playing', pos = 'v'))#verd
print(lemmantizer.lemmatize('playing', pos = 'n'))#noun
print(lemmantizer.lemmatize('playing', pos = 'a'))#adjective
print(lemmantizer.lemmatize('playing', pos = 'r'))#adverb
```

```python
# classifer the text 
import nltk.classify.util
from nltk.classify import NaiveBayesClassifier
from nltk.corpus import names
def f(word):
    return {'last_letter' : word[-1]}
#load data 

names = ([(name, 'male') for name in names.words('male.txt')] + [(name, 'female') for name in names.words('female.txt')] )

# training data 

featuresets = [(f(n),g) for (n,g) in names]
train_set = featuresets
classifier = nltk.NaiveBayesClassifier.train(train_set)
print(classifier.classify(f('lis')))
```
